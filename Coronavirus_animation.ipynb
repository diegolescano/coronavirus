```python
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
asia = ["Mainland China", "Thailand", "Japan", "South Korea", "Taiwan",
       "Macau", "Hong Kong", "Singapore", "Vietnam", "Nepal", "Malaysia", "Cambodia", "Sri Lanka", "United Arab Emirates", "Philippine", "India", "Iran"]
```

```python
europe = ['France', 'Germany', 'Finland', 'Italy', 'UK', 'Russia', 'Sweden', 'Spain',
       'Belgium']
```

```python
north_america = ['US', 'Canada']
```

```python
africa = ['Egypt']
```

```python
australia = ['Australia']
```

```python
def get_continent(country):
    if country in asia:
        return 'Asia'
    elif country in europe:
        return 'Europe'
    elif country in north_america:
        return 'North America'
    elif country in africa:
        return 'Africa'
    elif country in australia:
        return 'Australia'
    else:
        return 'Others'
```

```python
data['Continent'] = data['Country/Region'].apply(get_continent)
```

```python
#new_data_frame = pd.melt(data, id_vars=['Province/State','Country/Region','Lat','Long','Continent'], var_name='Date', value_name='Confirmed')
new_data_frame = pd.melt(data, id_vars=['Province/State','Country/Region','Lat','Long'], var_name='Date', value_name='Confirmed')
```

```python
new_data_frame.info()
```

```python
#from datetime import datetime
#new_data_frame['Date'] = datetime.strptime(new_data_frame['Date'], '%m/%d/%y')
#new_data_frame['Date'] = new_data_frame['Date'].dt.date
new_data_frame['Date'] = pd.to_datetime(new_data_frame['Date']).dt.strftime('%m/%d/%Y')
```

```python
new_data_frame.Date.unique()
```

```python
for i in new_data_frame.Date.unique():
    temp = new_data_frame[(new_data_frame["Date"] == i) & (new_data_frame["Confirmed"] != 0)]
    temp = temp.sort_values(by = 'Confirmed', ascending = False)

    fig = make_subplots(
        rows=2, cols=1,
        specs=[[{"type": "xy"}],
               [{"type": "scattergeo"}]],
    )
    fig.add_trace(go.Bar(x=temp['Country/Region'].value_counts().index,y=temp['Country/Region'].value_counts()), 
                  row=1, col=1)

    fig.add_trace(go.Scattergeo(
            lon = temp['Long'],
            lat = temp['Lat'],
            hovertext=temp[['Country/Region','Province/State']],
            text = temp['Province/State'],
            mode = 'markers',
            marker = dict(
            size = 8,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle',
            line = dict(
                width=.5,
                color='rgba(102, 102, 102)'
            ),
            colorscale = 'Blues',
            cmin = 0,
            color = 'red',
            )),row=2, col=1)

    fig.update_layout(height=500, width=1200, showlegend=False, title = '' + pd.to_datetime(i).strftime("%A %B %d, %Y"), 
                      titlefont = {"size": 36})

    fig.show()
    filename='Coronavirus_date'+str(i).replace('/','_')+'.png'
    fig.write_image(filename)

# Generate
#./magick convert -delay 80 Corona*.png coronavirus.gif
```
